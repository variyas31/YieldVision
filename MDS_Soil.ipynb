```python
path = '/Users/school/Desktop/Research_Paper_Material/DataFiles/' # put your working folder path here
fn = 'mds_TS_data.feather'
df =  pd.read_feather(path + fn, columns=None, use_threads=True)
```

## Another look by ploting data

```python
fig, ax = plt.subplots(figsize=[12,9])
sns.lineplot(data = df, x="daysaftersowing", y="yield", hue="name")
```
